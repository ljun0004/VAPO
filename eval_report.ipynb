```python
import numpy as np
from matplotlib import pyplot as plt
import os

eval_dir = "./cifar10_FM_mean_corr_cos_sc1_sl1_wc1_wl1_wn0.5_wt0.1_ts1e-5_Adam_silu/eval"

iter = []
incept_score=[]
fid = []
reports = []

for p in os.listdir(eval_dir):
    if "report" in p:
        report = np.load(f"{eval_dir}/{p}")
        incept_score=report['IS'].item()
        fid=report['fid'].item()
        ckpt=int(p.split(".npz")[0].split("_")[-1])
        iter=ckpt*10000
        reports.append((iter, fid, incept_score))
reports = sorted(reports, key=lambda x: x[0])
reports_arr = np.array(reports)

print(reports_arr[-25:, 0]/ 10000)
print(reports_arr[-25:, 0])
print(reports_arr[-25:, 1])
plt.plot(reports_arr[-25:, 0], reports_arr[-25:, 1], 'bo-')
```
